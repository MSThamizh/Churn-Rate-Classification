```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
# Read the data from the csv file
df = pd.read_csv(r"D:\Guvi\Projects\Churn Rate Classification\Dataset\churnrate_data.csv")
print(type(df))
df
```

```python
df.info()
```

```python
# Print the number of null values under each column
print("Null values per column:")
print(df.isna().sum())
```

```python
# Remove the columns with high number of null values
df.drop(columns=['credit_score','rewards_earned'],inplace=True)
```

```python
# Make the age attribute non-null
df = df[pd.notnull(df['age'])]
df['age']
# So the 4 rows will be deleted
```

```python
print("Null values per column finally")
print(df.isna().sum())
```

```python
df.info()
```

```python
df.columns
```

```python
# Remove columns that are mostly not useful for the classification task
df = df.drop(columns=['app_web_user','user'])
```

```python
# Performing encding on the attributes
# One-hot encoding
df = pd.get_dummies(df, dtype=int)
df.columns
```

```python
# Drop a  few more columns
df = df.drop(columns=['housing_na','zodiac_sign_na','payment_type_na'])
df.info()
```

```python
df
```

```python
# Split the data into train and test data
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['churn']),df['churn'],test_size=0.3)
```

```python
# Check for class imbalance
print("Count of each class in the training set:")
y_train.value_counts()
```

```python
# Resample to handle class imbalance

positive = y_train[y_train.values == 1]
negative = y_train[y_train.values == 0]
if len(positive) > len(negative):
    high = positive
    low = negative
else:
    high = negative
    low = positive

df_resample = resample(high, replace = False, n_samples = int(low.value_counts()))
```

```python
df_resample.info()
```

```python
# Standardize the data
sc_x = StandardScaler()
x_train = pd.DataFrame(sc_x.fit_transform(X_train))
x_test = pd.DataFrame(sc_x.fit_transform(X_test))

x_train.head()
```

```python
# Add Columns to New Dataframes
x_train.columns = X_train.columns.values
x_test.columns = X_test.columns.values

x_train.head()
```

```python
# Use Logistic Regression as classifier
model = LogisticRegression(class_weight = None, solver = 'newton-cg', max_iter = 10000)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

# Obtain the results of the classification
from sklearn.metrics import classification_report, accuracy_score
print("Accuracy with Logistic Regression Model")
print(accuracy_score(y_test,y_pred))

print("Classification Report with Logistic Regression Model")
print(classification_report(y_test,y_pred))
```

```python
# Use Random forest classifier
clf = RandomForestClassifier(n_estimators = 100)
# Train the model
clf.fit(x_train,y_train)
# Predict
y_pred = clf.predict(x_test)

# Evaluate Random forest classifier
print("Accuracy with Random forest classifier Model")
print(accuracy_score(y_test,y_pred))

print("Classification Report with Random forest classifier Model")
print(classification_report(y_test,y_pred))
```
